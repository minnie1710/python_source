```python
import pandas as pd

train = pd.read_csv("train.csv", parse_dates=["datetime"])
print(train.shape)
train.head()
```

```python
test = pd.read_csv("test.csv", parse_dates=["datetime"])
print(test.shape)
test.head()
```

```python
train.describe()
```

```python
test.describe()
```

```python
train.columns
```

```python
feature_names = ['season', 'holiday', 'workingday', 'weather', 'temp',
       'atemp', 'humidity', 'windspeed']
feature_names
train[feature_names]
```

```python
X_train = train[feature_names]

print(X_train.shape)
X_train.head()
```

```python
X_test = test[feature_names]

print(X_test.shape)
X_test.head()
```

```python
label_name = "count"
y_train = train[label_name]
train.columns
y_train.head()
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(max_depth=37)
model
```

```python
from sklearn.model_selection import cross_val_score

score1 = cross_val_score(model, X_train, y_train, cv=20,
                        scoring="neg_mean_absolute_error").mean()
score1
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print(predictions.shape)
predictions[0:10]
```

```python
submission =  pd.read_csv("sampleSubmission.csv", index_col="datetime")
submission
```

```python
submission["count"] = predictions
print(submission.shape)
submission.head()
```

```python
submission.to_csv("decision-tree1.csv")
```
